```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Data in the given format
data = (np.array([[-1.24614844, -2.27962969],
                 [-0.50766043, -0.56122324],
                 [-1.36801826, -2.19693941]]),
        np.array([0, 0, 1]))

# Unpack the data
X, y = data

# Train a logistic regression model
clf = LogisticRegression(solver='liblinear')
clf.fit(X, y)

# Plot the data points
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', color='blue')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', color='red')

# Plot the decision boundary
coef = clf.coef_[0]
intercept = clf.intercept_
x_values = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_values = (-intercept - coef[0] * x_values) / coef[1]
plt.plot(x_values, y_values, label='Decision Boundary', color='green')

# Add labels and legend
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# Show the plot
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Generate a list of 10 random 2D points
points = [(random.uniform(0, 10), random.uniform(0, 10)) for _ in range(10)]

# Generate a list of labels (0 or 1) for each point
labels = [random.choice([0, 1]) for _ in range(10)]

# Create the tuple with points and labels
data_tuple = (points, labels)

# Separate points based on labels
points_0 = [point for point, label in zip(data_tuple[0], data_tuple[1]) if label == 0]
points_1 = [point for point, label in zip(data_tuple[0], data_tuple[1]) if label == 1]

# Extract x and y coordinates for points
x_0, y_0 = zip(*points_0)
x_1, y_1 = zip(*points_1)

# Plot the points
plt.scatter(x_0, y_0, color='red', label='Label 0', marker='o')
plt.scatter(x_1, y_1, color='blue', label='Label 1', marker='^')

# Add labels to the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best')

# Show the plot
plt.show()
```
